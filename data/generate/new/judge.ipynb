```python
import pickle
import re
import random
def readPkl(rdAddr): 
    readList = []
    with open(rdAddr, 'rb') as f:
        readList=pickle.load(f)
    f.close()
    return readList

def savePKL(pklAddr, saveList):
    pk = open(pklAddr, 'wb')
    random.shuffle(saveList)
    random.shuffle(saveList)
    pickle.dump(saveList, pk)
    pk.close()
```
